```python
import numpy as np

def perceptron(X, y, epochs=1):
    w = np.zeros(X.shape[1])
    b = 0

    for _ in range(epochs):
        for i in range(len(X)):
            if y[i] * (np.dot(w, X[i]) + b) <= 0:
                w += y[i] * X[i]
                b += y[i]

    return w, b
```
