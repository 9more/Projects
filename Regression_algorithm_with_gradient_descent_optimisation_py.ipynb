```python
def metric(w ,b, ytest, Xtest, measure='mse'):
  if measure=='rmse':
    y=w@Xtest.T+b
    e=y-ytest
    rmse=np.sqrt(np.mean(e**2))
    return rmse
  elif measure=='mae':
    y=w@Xtest.T+b
    e=y-ytest
    mae=np.mean(np.abs(e))
    return mae
  else:
    y=w@Xtest.T+b
    e=y-ytest
    mse=np.mean(e**2)
    return mse
```
