```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
#%pip install mysqlclient

user_name, password = input('user_name:'), input('password:')
rds_endpoint, database = input('endpoint:'), input('database:')
db_type, port= 'mysql+pymysql','3306'
engine = create_engine(f'{db_type}://{user_name}:{password}@{rds_endpoint}:{port}/{database}')
query="""
        SELECT cp.NAME_OF_SCHOOL, cp.COMMUNITY_AREA_NAME, cp.`Elementary, Middle, or High School`,cp.City,State,cp.ZIP_Code,
        cp.SAFETY_SCORE, cp.Environment_Score, cp.AVERAGE_STUDENT_ATTENDANCE, cp.College_Eligibility__,cp.College_Enrollment_Rate__,
        cp.Graduation_Rate__, cr.CASE_NUMBER, cr.PRIMARY_TYPE, cr.`DESCRIPTION`, cr.DISTRICT, cr.WARD, cr.YEAR,
        cs.HARDSHIP_INDEX,cs.PER_CAPITA_INCOME, cs.PERCENT_AGED_25__WITHOUT_HIGH_SCHOOL_DIPLOMA, cs.PERCENT_AGED_16__UNEMPLOYED,
        cs.PERCENT_AGED_UNDER_18_OR_OVER_64, cs.PERCENT_HOUSEHOLDS_BELOW_POVERTY, cs.PERCENT_OF_HOUSING_CROWDED
        FROM my_schema.chicago_public_schools cp
        INNER JOIN chicago_crime cr
        ON cp.COMMUNITY_AREA_NUMBER = cr.COMMUNITY_AREA_NUMBER
        INNER JOIN chicago_socioeconomic_data cs
        ON cp.COMMUNITY_AREA_NUMBER = cs.COMMUNITY_AREA_NUMBER;
      """
data=pd.read_sql(query,engine)
```
